### Part I. Data Extraction from Bloomberg API

### Import necessary libraries

```python
# data manupulation libraries
import pandas as pd
import numpy as np

# bloomberg API querying libraries
import xbbg
from xbbg import blp
```

### Data Preparation

```python
# read 3000 tickers from bloomberg that I got by running EQS (equity screening function) to get all actively traded at US Exchanges
tickers = pd.read_csv('tickers.csv')
tickers
```

```python
# extract necassary historical financial information from bloomberg API for the tickers above
fields = ['ESG_SCORE', 'ESG_DISCLOSURE_SCORE', 'BETA_RAW_OVERRIDABLE','PE_RATIO','PX_TO_BOOK_RATIO', 'IS_EPS', 'RETURN_COM_EQY', 'TOT_DEBT_TO_TOT_EQY', 'BOOK_VAL_PER_SH', 'PROF_MARGIN','OPER_INC_GROWTH', 'RETURN_ON_ASSET']
stocks_data = pd.DataFrame()

for  ticker  in  tickers['Ticker']:
    hist_data = blp.bdh(tickers=ticker, flds = fields, Per='Y', start_date='2015-01-01', end_date='2022-12-31')    
    blank_data = blp.bdp(tickers=ticker, flds = ['GICS_SECTOR_NAME','CUR_MKT_CAP'])
    
    if len(hist_data)>0  and blank_data.shape == (1,2):        
        hist_data.columns = hist_data.columns.droplevel()
        hist_data['Ticker'] = ticker        
        hist_data['Industry'] = blank_data['gics_sector_name'][0]
        hist_data['MarketCap'] = blank_data['cur_mkt_cap'][0]
        stocks_data = pd.concat([hist_data, stocks_data], axis=0)
        
stocks_data_copy = stocks_data.copy()
stocks_data.to_csv('data_raw.csv')
stocks_data
```
